# Extraction Based Summarizer

Aim: Getting summary of the wikipedia

```python
import bs4 as bs
import urllib.request
import nltk
nltk.download('punkt')
scraped_data=urllib.request.urlopen('https://en.wikipedia.org/wiki/Iranian_cuisine')
data=scraped_data.read()
parsed_article=bs.BeautifulSoup(data,'lxml')
parsed_article.head()
```

```python
paragraph=parsed_article.find_all('p')
para_text=''
for i in paragraph:
    para_text+=i.text


para_text[:1000]
```

```python
import re
# Removing Square Brackets and Extra Spaces
para_text = re.sub(r'\[[0-9]*\]', ' ',para_text)
#any whitespace character \s+
para_text = re.sub(r'\s+', ' ', para_text)
para_text[:1000]
```

```python
# Removing special characters and digits
formatted_para_text = re.sub('[^a-zA-Z]', ' ', para_text )
#any whitespace character \s+
formatted_para_text = re.sub(r'\s+', ' ', formatted_para_text)
formatted_para_text[:1000]
```

```python
sentence_list=nltk.sent_tokenize(para_text)
sentence_list[:5]
```

```python
word_frequency={}
stopwords=nltk.corpus.stopwords.words('english')
for word in nltk.word_tokenize(formatted_para_text):
    if word not in stopwords:
        if word not in word_frequency:
            word_frequency[word]=1
        else:
            word_frequency[word]+=1


word_frequency
```

```python
maxvalue=max(word_frequency.values())
maxvalue
```

```python
for word in word_frequency:
    word_frequency[word]=word_frequency[word]/maxvalue
word_frequency
```

```python
sentence_score={}
for sent in nltk.sent_tokenize(para_text):
    for word in nltk.word_tokenize(sent.lower()):
        if word in word_frequency.keys():
            if sent not in sentence_score.keys():
                sentence_score[sent]=word_frequency[word]
            else:
                sentence_score[sent]+=word_frequency[word]

sentence_score
```

```python
import heapq
summary_sentences = heapq.nlargest(2, sentence_score, key=sentence_score.get)
summary = ' '.join(summary_sentences)
summary
```
